```python
amostra = pnad.loc[ (pnad.Anos >= 18) & (pnad.Anos <= 80) & (pnad.Salario > 0) ]
```

```python
amostra.describe()
```

```python
plt.scatter(amostra.Anos, amostra.Salario)
plt.title("Relação entre Anos e Salário")
plt.xlabel("Anos de Experiência")
plt.ylabel("Salário")
plt.show()
```

```python
dado = amostra.groupby('Anos').agg({'Salario' : 'mean'})
dado.reset_index(inplace=True)
dado
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x=dado.Anos, y=dado.Salario)
plt.xlabel('Idade')
plt.ylabel('Média da renda')
plt.show()
```

```python
import statsmodels.formula.api as smf
```

```python
modelo1 = smf.ols(formula='Salario ~ Anos', data=dado).fit()
print(modelo1.summary())
```

```python
yhat1 = modelo1.predict(dado.Anos)

plt.figure(figsize=(10,5))
plt.scatter(x=dado.Anos, y=dado.Salario)
plt.plot(dado.Anos, yhat1, c='red')
plt.xlabel('Idade')
plt.ylabel('Média da renda')
plt.show()
```
